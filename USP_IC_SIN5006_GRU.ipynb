# üöÄ Pipeline - Modelo GRU - Projeto USP-IC-SIN5006
Este notebook executa o modelo GRU sobre todas as bases localizadas na pasta `datasets` do Google Drive.
Os resultados (previs√µes, gr√°ficos e m√©tricas) ser√£o salvos automaticamente na pasta `outputs`.

```python

# üîß Instala√ß√£o de pacotes necess√°rios
!pip install tensorflow scikit-learn pandas matplotlib seaborn plotly --quiet

# üîó Montar Google Drive
from google.colab import drive
drive.mount('/content/drive')

# üóÇÔ∏è Definir os caminhos das pastas
base_path = '/content/drive/MyDrive/USP-IC-SIN5006/'
datasets_path = base_path + 'datasets/'
outputs_path = base_path + 'outputs/'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM, GRU
import os
import glob

datasets = glob.glob(datasets_path + '*.csv')
print(f'üóÇÔ∏è Bases encontradas: {datasets}')

def create_sequences(data, look_back=30):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:(i + look_back)])
        y.append(data[i + look_back])
    return np.array(X), np.array(y)

def build_model(input_shape):
    model = Sequential()
    if 'GRU' == 'LSTM':
        model.add(LSTM(50, activation='tanh', input_shape=input_shape))
    elif 'GRU' == 'GRU':
        model.add(GRU(50, activation='tanh', input_shape=input_shape))
    elif 'GRU' == 'MLP':
        model.add(Dense(64, activation='relu', input_shape=(input_shape[0],)))
        model.add(Dense(32, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred) ** 0.5
    r2 = r2_score(y_true, y_pred)
    return {'MAE': mae, 'RMSE': rmse, 'R2': r2}

results = []
look_back = 30

for dataset_path in datasets:
    dataset_name = os.path.basename(dataset_path).replace('.csv','')
    print(f"üìà Processando dataset: {dataset_name}")

    df = pd.read_csv(dataset_path)

    # Drop the date column and then select all remaining columns
    if 'date_m' in df.columns:
        df_numerical = df.drop(columns=['date_m'])
    else:
        df_numerical = df.copy() # If 'date_m' is not present, use the whole dataframe

    # Ensure there are numerical columns to process
    if df_numerical.shape[1] == 0:
        print(f"‚ö†Ô∏è Skipping dataset {dataset_name}: No numerical columns found after dropping date column.")
        continue

    values = df_numerical.values

    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled = scaler.fit_transform(values)

    # Since we are now using multiple columns for the input sequence,
    # the target variable 'y' should be based on the column you want to predict.
    # For now, I'll assume you want to predict the first numerical column (index 0 of values).
    # You might need to adjust this based on your specific prediction target.
    X, y = [], []
    # Ensure there are enough data points for the look_back window after handling potential NaNs
    if len(scaled) > look_back:
        for i in range(len(scaled) - look_back):
            X.append(scaled[i:(i + look_back)])
            y.append(scaled[i + look_back, 0]) # Predicting the first numerical column


    X = np.array(X)
    y = np.array(y)

    # Skip processing this dataset if there are not enough data points after creating sequences
    if len(X) == 0:
        print(f"‚ö†Ô∏è Skipping dataset {dataset_name}: Not enough data points for look_back window after processing.")
        continue


    X_model = X.reshape((X.shape[0], X.shape[1], X.shape[2])) if 'GRU' in ['LSTM', 'GRU'] else X.reshape((X.shape[0], X.shape[1] * X.shape[2]))


    input_shape = (X.shape[1], X.shape[2]) if 'GRU' in ['LSTM', 'GRU'] else (X.shape[1] * X.shape[2],)

    model = build_model(input_shape)

    history = model.fit(
        X_model,
        y,
        epochs=50,
        batch_size=16,
        verbose=0
    )

    y_pred = model.predict(X_model)

    # Inverse transform y_true and y_pred.
    # Since y is now a 1D array (predicting a single column), we need to handle the inverse transform carefully.
    # We need to create a dummy array with the same number of features as the original 'values' to use the scaler.
    y_true_inv = scaler.inverse_transform(np.concatenate((y.reshape(-1, 1), np.zeros((len(y), values.shape[1] - 1))), axis=1))[:, 0]
    y_pred_inv = scaler.inverse_transform(np.concatenate((y_pred.reshape(-1, 1), np.zeros((len(y_pred), values.shape[1] - 1))), axis=1))[:, 0]


    metrics = evaluate_model(y_true_inv, y_pred_inv)
    metrics['Dataset'] = dataset_name
    metrics['Model'] = 'GRU'
    results.append(metrics)

    plt.figure(figsize=(10,5))
    plt.plot(y_true_inv, label='Real')
    plt.plot(y_pred_inv, label='Previsto')
    plt.title(f'Previs√£o - {dataset_name} - GRU')
    plt.legend()
    plot_file = outputs_path + f'plot_{dataset_name}_GRU.png'
    plt.savefig(plot_file)
    plt.show()

    df_out = pd.DataFrame({
        'Real': y_true_inv.flatten(),
        'Previsto': y_pred_inv.flatten()
    })
    output_name = outputs_path + f'resultado_{dataset_name}_GRU.csv'
    df_out.to_csv(output_name, index=False)
    print(f"‚úÖ Resultado salvo: {output_name}")
    print(f"üñºÔ∏è Gr√°fico salvo: {plot_file}")

df_results = pd.DataFrame(results)
print(df_results)
df_results.to_csv(outputs_path + 'Resultados_Consolidados_GRU.csv', index=False)
print("üìÑ Resultados consolidados salvos em 'Resultados_Consolidados_GRU.csv'")
```
