# üöÄ Pipeline - Modelo LSTM - Projeto USP-IC-SIN5006
Este notebook executa o modelo LSTM sobre todas as bases localizadas na pasta `datasets` do Google Drive.
Os resultados (previs√µes, gr√°ficos e m√©tricas) ser√£o salvos automaticamente na pasta `outputs`.

```python

# üîß Instala√ß√£o de pacotes necess√°rios
!pip install tensorflow scikit-learn pandas matplotlib seaborn plotly --quiet

# üîó Montar Google Drive
from google.colab import drive
drive.mount('/content/drive')

# üóÇÔ∏è Definir os caminhos das pastas
base_path = '/content/drive/MyDrive/USP-IC-SIN5006/'
datasets_path = base_path + 'datasets/'
outputs_path = base_path + 'outputs/'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM, GRU
import os
import glob

datasets = glob.glob(datasets_path + '*.csv')
print(f'üóÇÔ∏è Bases encontradas: {datasets}')

def create_sequences(data, look_back=30):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:(i + look_back)])
        y.append(data[i + look_back])
    return np.array(X), np.array(y)

def build_model(input_shape):
    model = Sequential()
    if 'LSTM' == 'LSTM':
        model.add(LSTM(50, activation='tanh', input_shape=input_shape))
    elif 'LSTM' == 'GRU':
        model.add(GRU(50, activation='tanh', input_shape=input_shape))
    elif 'LSTM' == 'MLP':
        model.add(Dense(64, activation='relu', input_shape=(input_shape[0],)))
        model.add(Dense(32, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred) ** 0.5
    r2 = r2_score(y_true, y_pred)
    return {'MAE': mae, 'RMSE': rmse, 'R2': r2}

results = []
look_back = 30

for dataset_path in datasets:
    dataset_name = os.path.basename(dataset_path).replace('.csv','')
    print(f"üìà Processando dataset: {dataset_name}")

    df = pd.read_csv(dataset_path)
    # Select only numerical columns for scaling
    values = df.select_dtypes(include=np.number).values

    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled = scaler.fit_transform(values)

    X, y = create_sequences(scaled, look_back)

    X_model = X.reshape((X.shape[0], X.shape[1], X.shape[2])) if 'LSTM' in ['LSTM', 'GRU'] else X.reshape((X.shape[0], X.shape[1]))

    input_shape = (X.shape[1], X.shape[2]) if 'LSTM' in ['LSTM', 'GRU'] else (X.shape[1],)
    model = build_model(input_shape)

    history = model.fit(
        X_model,
        y,
        epochs=50,
        batch_size=16,
        verbose=0
    )

    y_pred = model.predict(X_model)

    # Reshape y_pred to have the same number of features as scaled data
    y_pred_reshaped = np.repeat(y_pred, scaled.shape[1], axis=-1)

    y_true_inv = scaler.inverse_transform(y)
    y_pred_inv = scaler.inverse_transform(y_pred_reshaped)

    metrics = evaluate_model(y_true_inv[:, 0], y_pred_inv[:, 0]) # Evaluate using only the first column
    metrics['Dataset'] = dataset_name
    metrics['Model'] = 'LSTM'
    results.append(metrics)

    plt.figure(figsize=(10,5))
    plt.plot(y_true_inv[:, 0], label='Real') # Plot the first column of y_true_inv
    plt.plot(y_pred_inv[:, 0], label='Previsto') # Plot the first column of y_pred_inv
    plt.title(f'Previs√£o - {dataset_name} - LSTM')
    plt.legend()
    plot_file = outputs_path + f'plot_{dataset_name}_LSTM.png'
    plt.savefig(plot_file)
    plt.show()

    df_out = pd.DataFrame({
        'Real': y_true_inv[:, 0].flatten(), # Select and flatten the first column of y_true_inv
        'Previsto': y_pred_inv[:, 0].flatten() # Select and flatten the first column of y_pred_inv
    })
    output_name = outputs_path + f'resultado_{dataset_name}_LSTM.csv'
    df_out.to_csv(output_name, index=False)
    print(f"‚úÖ Resultado salvo: {output_name}")
    print(f"üñºÔ∏è Gr√°fico salvo: {plot_file}")

df_results = pd.DataFrame(results)
print(df_results)
df_results.to_csv(outputs_path + 'Resultados_Consolidados_LSTM.csv', index=False)
print("üìÑ Resultados consolidados salvos em 'Resultados_Consolidados_LSTM.csv'")
```
